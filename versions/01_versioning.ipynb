## Let's load the data

```python
# We first define dtype_dict ie the types of column but crawl_date as we will parse this column
dtype_dict={'domain': str, 'url': str, 'content':str,'language':str}
# We load the csv as a dataframe. Remove (or add) the nrows=100000 if you want to load the whole dataframe or just a part.
web_pages =  pd.read_csv('generated/web_pages_en_deduplicated.csv', usecols=['crawl_date','date','domain','url','content'], dtype=dtype_dict, parse_dates=['crawl_date','date']) #uncomment this line and comment the next one if you would prefer to play with the first x rows to spare memory and time.
web_pages.sort_values(by='crawl_date')
```

```python
#let's group by URL to see the number of versions we need to look at.
web_pages_byURL = web_pages.groupby(['url']).size()
```
